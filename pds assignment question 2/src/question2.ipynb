```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

```python
student= pd.read_csv('../clean_data/NewStudentsPerformance.csv')
```

```python
sns.barplot(x = 'parental_level_of_education', y = 'writing_score',data = student)
 
```

```python
student.groupby(['gender']).sum().plot(
    kind='pie', y='reading_score', autopct='%1.0f%%')
```

```python
plt.figure(figsize=(20,10))
sns.countplot(data=student,x=student['math_score'],hue=student['gender'])
plt.title("male and female comparison of marks")
plt.show()
```

```python
plt.figure(figsize = (14, 8))
sns.set_style('white')



plt.scatter(data = student.head(100), x = "reading_score", y = "writing_score", )

plt.xlabel('parental_level_of_education', fontsize = 13)
plt.ylabel('race/ethnicity', fontsize = 13)

plt.title('Scatter Plot 1', fontsize = 14, fontweight = 'bold')

sns.despine()

plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(student['math_score'])
plt.title(('Countplot of Math Score'))
```
